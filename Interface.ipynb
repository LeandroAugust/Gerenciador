```python
import tkinter as tk

class Homepage:
    def __init__(self):
        self.inicio = tk.Tk()

        self.inicio.title("Pomodoro")
        self.inicio.geometry("720x580+100+50")

        self.current_language = 'pt'
        self.texts = {
            'pt': {
                'title': "Pomodoro",
                'ferramentas': "Ferramentas",
                'linguagens': "Linguagens",
                'tema': "Tema",
                'claro': "Claro",
                'escuro': "Escuro",
                'ajuda': "Ajuda",
                'sobre': "Sobre",
                'como_usar': "Como Usar?",
                'sobre_texto': "Gestão de Tempo pelo método pomodoro"
            },
            'en': {
                'title': "Pomodoro",
                'ferramentas': "Tools",
                'linguagens': "Languages",
                'tema': "Theme",
                'claro': "Light",
                'escuro': "Dark",
                'ajuda': "Help",
                'sobre': "About",
                'como_usar': "How to Use?",
                'sobre_texto': "Time Management by the Pomodoro Method"
            },
            'ja': {
                'title': "ポモドーロ",
                'ferramentas': "ツール",
                'linguagens': "言語",
                'tema': "テーマ",
                'claro': "ライト",
                'escuro': "ダーク",
                'ajuda': "ヘルプ",
                'sobre': "約",
                'como_usar': "使い方は？",
                'sobre_texto': "ポモドーロ方式による時間管理"
            },
            'de': {
                'title': "Pomodoro",
                'ferramentas': "Werkzeuge",
                'linguagens': "Sprachen",
                'tema': "Thema",
                'claro': "Hell",
                'escuro': "Dunkel",
                'ajuda': "Hilfe",
                'sobre': "Über",
                'como_usar': "Wie benutzt man?",
                'sobre_texto': "Zeitmanagement mit der Pomodoro-Methode"
            },
            'la': {
                'title': "Pomodoro",
                'ferramentas': "Instrumenta",
                'linguagens': "Linguarum",
                'tema': "Thema",
                'claro': "Luce",
                'escuro': "Nigrum",
                'ajuda': "Auxilium",
                'sobre': "De",
                'como_usar': "Quomodo Utere?",
                'sobre_texto': "Tempus Management per methodum Pomodoro"
            },
            'es': {
                'title': "Pomodoro",
                'ferramentas': "Herramientas",
                'linguagens': "Idiomas",
                'tema': "Tema",
                'claro': "Claro",
                'escuro': "Oscuro",
                'ajuda': "Ayuda",
                'sobre': "Acerca de",
                'como_usar': "¿Cómo usar?",
                'sobre_texto': "Gestión del tiempo por el método Pomodoro"
            },
            'zh': {
                'title': "番茄钟",
                'ferramentas': "工具",
                'linguagens': "语言",
                'tema': "主题",
                'claro': "光",
                'escuro': "暗",
                'ajuda': "帮助",
                'sobre': "关于",
                'como_usar': "如何使用?",
                'sobre_texto': "通过番茄工作法进行时间管理"
            },
            'ko': {
                'title': "뽀모도로",
                'ferramentas': "도구",
                'linguagens': "언어",
                'tema': "주제",
                'claro': "밝은",
                'escuro': "어두운",
                'ajuda': "도움",
                'sobre': "약",
                'como_usar': "사용 방법?",
                'sobre_texto': "뽀모도로 방법으로 시간 관리"
            },
            'ar': {
                'title': "بومودورو",
                'ferramentas': "أدوات",
                'linguagens': "اللغات",
                'tema': "موضوع",
                'claro': "فاتح",
                'escuro': "داكن",
                'ajuda': "مساعدة",
                'sobre': "حول",
                'como_usar': "كيفية الاستخدام؟",
                'sobre_texto': "إدارة الوقت باستخدام طريقة بومودورو"
            },
            'ru': {
                'title': "Помидор",
                'ferramentas': "Инструменты",
                'linguagens': "Языки",
                'tema': "Тема",
                'claro': "Светлый",
                'escuro': "Темный",
                'ajuda': "Помощь",
                'sobre': "О",
                'como_usar': "Как использовать?",
                'sobre_texto': "Управление временем по методу Помидора"
            },
            'pl': {
                'title': "Pomodoro",
                'ferramentas': "Narzędzia",
                'linguagens': "Języki",
                'tema': "Temat",
                'claro': "Jasny",
                'escuro': "Ciemny",
                'ajuda': "Pomoc",
                'sobre': "O",
                'como_usar': "Jak używać?",
                'sobre_texto': "Zarządzanie czasem metodą Pomodoro"
            },
            'fr': {
                'title': "Pomodoro",
                'ferramentas': "Outils",
                'linguagens': "Langues",
                'tema': "Thème",
                'claro': "Clair",
                'escuro': "Sombre",
                'ajuda': "Aide",
                'sobre': "À propos",
                'como_usar': "Comment utiliser?",
                'sobre_texto': "Gestion du temps par la méthode Pomodoro"
            }
        }

        self.create_menu()
        self.create_widgets()
        self.update_texts()
        self.light_mode()

        self.inicio.mainloop()

    def create_menu(self):
        self.top_frame = tk.Frame(self.inicio, height=30)
        self.top_frame.pack(fill='x')

        self.menubar = tk.Menu(self.inicio, tearoff=False)
        self.inicio.config(menu=self.menubar)

        self.MENUferramentas = tk.Menu(self.menubar, tearoff=False)
        self.menubar.add_cascade(label="", menu=self.MENUferramentas)

        self.menulang = tk.Menu(self.MENUferramentas, tearoff=False)
        self.menulang.add_command(label="Português", command=lambda: self.change_language('pt'))
        self.menulang.add_command(label="English", command=lambda: self.change_language('en'))
        self.menulang.add_command(label="日本語", command=lambda: self.change_language('ja'))
        self.menulang.add_command(label="Deutsch", command=lambda: self.change_language('de'))
        self.menulang.add_command(label="Latina", command=lambda: self.change_language('la'))
        self.menulang.add_command(label="Español", command=lambda: self.change_language('es'))
        self.menulang.add_command(label="中文", command=lambda: self.change_language('zh'))
        self.menulang.add_command(label="한국어", command=lambda: self.change_language('ko'))
        self.menulang.add_command(label="العربية", command=lambda: self.change_language('ar'))
        self.menulang.add_command(label="Русский", command=lambda: self.change_language('ru'))
        self.menulang.add_command(label="Polski", command=lambda: self.change_language('pl'))
        self.menulang.add_command(label="Français", command=lambda: self.change_language('fr'))
        self.MENUferramentas.add_cascade(label="", menu=self.menulang)

        self.menuamb = tk.Menu(self.MENUferramentas, tearoff=False)
        self.theme_var = tk.StringVar(value="light")
        self.menuamb.add_radiobutton(label="", variable=self.theme_var, value="light", command=self.light_mode)
        self.menuamb.add_radiobutton(label="", variable=self.theme_var, value="dark", command=self.dark_mode)
        self.MENUferramentas.add_cascade(label="", menu=self.menuamb)

        self.MENUajuda = tk.Menu(self.menubar, tearoff=False)
        self.MENUajuda.add_command(label="", command=self.sobre)
        self.MENUajuda.add_command(label="", command=lambda: None)
        self.menubar.add_cascade(label="", menu=self.MENUajuda)

    def create_widgets(self):
        self.label = tk.Label(self.inicio, text="Pomodoro")
        self.label.pack()

        self.canvas = tk.Canvas(self.inicio, width=200, height=200)
        self.canvas.pack()
        self.draw_graphics()

    def draw_graphics(self):
        self.canvas.create_oval(50, 50, 150, 150, fill="purple")
        self.canvas.create_line(0, 0, 200, 200, fill="red", width=2)
        self.canvas.create_line(200, 0, 0, 200, fill="red", width=2)

    def update_texts(self):
        texts = self.texts[self.current_language]
        self.inicio.title(texts['title'])
        self.menubar.entryconfig(0, label=texts['ferramentas'])
        self.MENUferramentas.entryconfig(0, label=texts['linguagens'])
        self.MENUferramentas.entryconfig(1, label=texts['tema'])
        self.menuamb.entryconfig(0, label=texts['claro'])
        self.menuamb.entryconfig(1, label=texts['escuro'])
        self.menubar.entryconfig(1, label=texts['ajuda'])
        self.MENUajuda.entryconfig(0, label=texts['sobre'])
        self.MENUajuda.entryconfig(1, label=texts['como_usar'])

    def change_language(self, language):
        self.current_language = language
        self.update_texts()
        self.apply_theme()

    def dark_mode(self):
        self.bg_color = "#2E2E2E"
        self.fg_color = "white"
        self.apply_theme()

    def light_mode(self):
        self.bg_color = "light gray"
        self.fg_color = "black"
        self.apply_theme()

    def apply_theme(self):
        self.inicio.configure(bg=self.bg_color)
        self.top_frame.configure(bg=self.bg_color)
        self._update_menu_bg(self.menubar, self.bg_color, self.fg_color)
        self._update_menu_bg(self.MENUferramentas, self.bg_color, self.fg_color)
        self._update_menu_bg(self.menulang, self.bg_color, self.fg_color)
        self._update_menu_bg(self.menuamb, self.bg_color, self.fg_color)
        self._update_menu_bg(self.MENUajuda, self.bg_color, self.fg_color)

        self.label.configure(bg=self.bg_color, fg=self.fg_color)
        self.canvas.configure(bg=self.bg_color)

    def _update_menu_bg(self, menu, bg, fg):
        menu.configure(bg=bg, fg=fg)
        for index in range(menu.index("end") + 1):
            menu.entryconfig(index, background=bg, foreground=fg)

    def sobre(self):
        root = tk.Toplevel(self.inicio)
        root.geometry("240x110+75+75")
        root.title(self.texts[self.current_language]['sobre'])

        texto = self.texts[self.current_language]['sobre_texto']
        text_on_label = tk.Label(root, text=texto, bg=self.bg_color, fg=self.fg_color)
        text_on_label.pack()

        lb2 = tk.Label(root, text="", bg=self.bg_color, fg=self.fg_color)
        lb2.pack()

Homepage()
```

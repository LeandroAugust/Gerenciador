```python
import tkinter as tk

class Homepage:
    def __init__(self):
        self.inicio = tk.Tk()

        self.inicio.title("Pomodoro")
        self.inicio.geometry("720x580+100+50")
        self.inicio['bg'] = '#F2F2F2'

        self.current_language = 'pt'
        self.texts = {
            'pt': {
                'title': "Pomodoro",
                'ferramentas': "Ferramentas",
                'linguagens': "Linguagens",
                'tema': "Tema",
                'claro': "Claro",
                'escuro': "Escuro",
                'ajuda': "Ajuda",
                'sobre': "Sobre",
                'como_usar': "Como Usar?",
                'sobre_texto': "Gestão de Tempo pelo método pomodoro"
            },
            'en': {
                'title': "Pomodoro",
                'ferramentas': "Tools",
                'linguagens': "Languages",
                'tema': "Theme",
                'claro': "Light",
                'escuro': "Dark",
                'ajuda': "Help",
                'sobre': "About",
                'como_usar': "How to Use?",
                'sobre_texto': "Time Management by the Pomodoro Method"
            }
        }

        self.create_menu()
        self.update_texts()

        self.inicio.mainloop()

    def create_menu(self):
        self.top_frame = tk.Frame(self.inicio, height=30, bg='#F2F2F2')
        self.top_frame.pack(fill='x')

        self.menubar = tk.Menu(self.inicio, tearoff=False)
        self.inicio.config(menu=self.menubar)

        self.MENUferramentas = tk.Menu(self.menubar, tearoff=False)
        self.menubar.add_cascade(label="", menu=self.MENUferramentas)

        self.menulang = tk.Menu(self.MENUferramentas, tearoff=False)
        self.menulang.add_command(label="Português", command=lambda: self.change_language('pt'))
        self.menulang.add_command(label="English", command=lambda: self.change_language('en'))
        self.MENUferramentas.add_cascade(label="", menu=self.menulang)

        self.menuamb = tk.Menu(self.MENUferramentas, tearoff=False)
        self.theme_var = tk.StringVar(value="light")
        self.menuamb.add_radiobutton(label="", variable=self.theme_var, value="light", command=self.light_mode)
        self.menuamb.add_radiobutton(label="", variable=self.theme_var, value="dark", command=self.dark_mode)
        self.MENUferramentas.add_cascade(label="", menu=self.menuamb)

        self.MENUajuda = tk.Menu(self.menubar, tearoff=False)
        self.MENUajuda.add_command(label="", command=self.sobre)
        self.MENUajuda.add_command(label="", command=lambda: None)
        self.menubar.add_cascade(label="", menu=self.MENUajuda)

    def update_texts(self):
        texts = self.texts[self.current_language]
        self.inicio.title(texts['title'])
        self.menubar.entryconfig(0, label=texts['ferramentas'])
        self.MENUferramentas.entryconfig(0, label=texts['linguagens'])
        self.MENUferramentas.entryconfig(1, label=texts['tema'])
        self.menuamb.entryconfig(0, label=texts['claro'])
        self.menuamb.entryconfig(1, label=texts['escuro'])
        self.menubar.entryconfig(1, label=texts['ajuda'])
        self.MENUajuda.entryconfig(0, label=texts['sobre'])
        self.MENUajuda.entryconfig(1, label=texts['como_usar'])

    def change_language(self, language):
        self.current_language = language
        self.update_texts()

    def dark_mode(self):
        self.inicio.configure(bg="#2E2E2E")
        self.top_frame.configure(bg="#2E2E2E")
        self._update_menu_bg(self.menubar, "#2E2E2E", "white")
        self._update_menu_bg(self.MENUferramentas, "#2E2E2E", "white")
        self._update_menu_bg(self.menulang, "#2E2E2E", "white")
        self._update_menu_bg(self.menuamb, "#2E2E2E", "white")
        self._update_menu_bg(self.MENUajuda, "#2E2E2E", "white")

    def light_mode(self):
        self.inicio.configure(bg="light gray")
        self.top_frame.configure(bg="light gray")
        self._update_menu_bg(self.menubar, "light gray", "black")
        self._update_menu_bg(self.MENUferramentas, "light gray", "black")
        self._update_menu_bg(self.menulang, "light gray", "black")
        self._update_menu_bg(self.menuamb, "light gray", "black")
        self._update_menu_bg(self.MENUajuda, "light gray", "black")

    def _update_menu_bg(self, menu, bg, fg):
        menu.configure(bg=bg, fg=fg)
        for index in range(menu.index("end") + 1):
            menu.entryconfig(index, background=bg, foreground=fg)

    def sobre(self):
        root = tk.Toplevel(self.inicio)
        root.geometry("240x110+75+75")
        root.title(self.texts[self.current_language]['sobre'])

        texto = self.texts[self.current_language]['sobre_texto']
        text_on_label = tk.Label(root, text=texto)
        text_on_label.pack()

        lb2 = tk.Label(root, text="")
        lb2.pack()

Homepage()
```
